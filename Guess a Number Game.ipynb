```python
# Palindrome of a number

num = 1331
temp = num
save = 0
while num!=0:
  rem = num%10
  save = save*10+rem
  num = num // 10

  if temp==save:
    print('Palindrom Number')
  else:
    print('Not a Palindrome')
```
